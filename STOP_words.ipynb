```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp('We just opened our wings, the flying part is coming soon')
```

```python
for token in doc:
  if token.is_stop:
    print(token)
```

```python
def preprocess(text):
  doc = nlp(text)
  no_stop = [token.text for token in doc if not token.is_stop and not token.is_punct]
  return " ".join(no_stop)
```

```python
text = 'We just opened our wings, the flying part is coming soon'
```

```python
preprocess(text)
```
